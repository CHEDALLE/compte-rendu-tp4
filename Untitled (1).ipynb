```python
# Module : signal_module.py
import numpy as np
import matplotlib.pyplot as plt

class Signal:
    def __init__(self, nbr, tab):
        self.nbr = nbr
        self.tab = tab

    def echantillons(self):
        return self.nbr

    def moyenne(self):
        return np.mean(self.tab)

    def correlation(self):
        return np.corrcoef(self.tab[:-1], self.tab[1:])[0, 1]

    def display(self):
        print(f"Nombre d'échantillons: {self.nbr}")
        plt.plot(self.tab)
        plt.show()

class Aleatoire(Signal):
    def __init__(self, nbr, sigma, mean):
        super().__init__(nbr, [])
        self.sigma = sigma
        self.mean = mean
        self.init_alea()

    def init_alea(self):
        self.tab = np.random.randn(self.nbr) * self.sigma + self.mean

    def correlation(self):
        print("Classe Aléatoire")
        return np.power(self.tab, 2)

class Deterministe(Signal):
    def __init__(self, nbr, amplitude):
        super().__init__(nbr, [])
        self.amplitude = amplitude
        self.valeurs()

    def valeurs(self):
        self.tab[:self.nbr//2] = np.arange(0, self.nbr//2) * self.amplitude
        self.tab[self.nbr//2:] = -self.amplitude * np.ones(self.nbr//2)

def addition(objet1, objet2):
    new_objet = Deterministe(objet1.nbr, 0)
    new_objet.tab = objet1.tab + objet2.tab
    return new_objet

# Programme principal
if __name__ == "__main__":
    aleatoire_obj = Aleatoire(20, 2, 1)
    deterministe_obj = Deterministe(20, 3)

    print("Aléatoire:")
    print(f"Nombre d'échantillons: {aleatoire_obj.echantillons()}")
    print(f"Moyenne: {aleatoire_obj.moyenne()}")
    print(f"Corrélation: {aleatoire_obj.correlation()}")
    aleatoire_obj.display()

    print("\nDéterministe:")
    print(f"Nombre d'échantillons: {deterministe_obj.echantillons()}")
    print(f"Moyenne: {deterministe_obj.moyenne()}")
    print(f"Corrélation: {deterministe_obj.correlation()}")
    deterministe_obj.display()

    aleatoire_obj.correlation()

    resultat = addition(aleatoire_obj, deterministe_obj)
    print("\nRésultat de l'addition:")
    print(f"Nombre d'échantillons: {resultat.echantillons()}")
    print(f"Moyenne: {resultat.moyenne()}")
    print(f"Corrélation: {resultat.correlation()}")
    resultat.display()
```

```python
# Module : employe_module.py
class Employe:
    def __init__(self, nom, identifiant, pin=500):
        self.nom = nom
        self.ID = identifiant
        self.Pin = pin

    def calcul_prime(self, a):
        return self.Pin * 2 if a else self.Pin * 0.5

    def salaire(self, prime):
        return self.Pin * prime

    def affiche(self, a):
        prime = self.calcul_prime(a)
        salary = self.salaire(prime)
        print(f"Nom: {self.nom}, ID: {self.ID}, Point_indiciaire: {self.Pin}")
        print(f"Salaire: {salary}, Statut: Employé")


class Ingenieur(Employe):
    def __init__(self, nom, identifiant, etat, pin=500):
        super().__init__(nom, identifiant, pin)
        self.Etat = etat

    def salaire(self, prime):
        multiplicateur = 4 if self.Etat == "stagiaire" else 6
        return self.Pin * multiplicateur * prime

    def affiche(self, a):
        prime = self.calcul_prime(a)
        salary = self.salaire(prime)
        print(f"Nom: {self.nom}, ID: {self.ID}, Point_indiciaire: {self.Pin}")
        print(f"Salaire: {salary}, Statut: Ingenieur {self.Etat}")


class Technicien(Employe):
    def salaire(self, prime):
        return 2 * self.Pin * prime

    def affiche(self, a):
        prime = self.calcul_prime(a)
        salary = self.salaire(prime)
        print(f"Nom: {self.nom}, ID: {self.ID}, Point_indiciaire: {self.Pin}")
        print(f"Salaire: {salary}, Statut: Technicien")


# Programme principal
ingenieur1 = Ingenieur("Jean", 1, "stagiaire")
technicien1 = Technicien("Alice", 2)

ingenieur1.affiche(True)
technicien1.affiche(False)
```
